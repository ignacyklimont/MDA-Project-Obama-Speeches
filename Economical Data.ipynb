```python
import csv
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt
import datetime as dt
import matplotlib.dates as mdates
import numpy as np
```

```python
with open('speeches_clean.csv', newline='') as file:
    reader = csv.reader(file)
    speeches = list(reader)
    
df = pd.DataFrame(speeches)
df.columns = ['Date', 'Title', 'Text']
```

```python
gdp = pd.read_csv("MDA-Project-Obama-Speeches\\US_Economical_Data\\GDP.csv", sep=";", decimal=",")
homeprices = pd.read_csv("MDA-Project-Obama-Speeches\\US_Economical_Data\\HOMEPRICES.csv", sep=";", decimal=",")
sp_500 = pd.read_csv("MDA-Project-Obama-Speeches\\US_Economical_Data\\INDEX_US_S&P.csv", sep=";", decimal=",")
unrate = pd.read_csv("MDA-Project-Obama-Speeches\\US_Economical_Data\\UNRATE.csv", sep=";", decimal=",")

x = [dt.datetime.strptime(d,'%d/%m/%Y').date() for d in gdp.iloc[:,0]]
x2 = [dt.datetime.strptime(d,'%d/%m/%Y').date() for d in homeprices.iloc[:,0]]
x3 = [dt.datetime.strptime(d,'%d/%m/%Y').date() for d in sp_500.iloc[:,0]]
x4 = [dt.datetime.strptime(d,'%d/%m/%Y').date() for d in unrate.iloc[:,0]]
```

```python
fig, axs = plt.subplots(2,2, figsize = (16, 10), sharex=True)  # Create a figure containing a single axes.
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%d/%m/%Y'))
plt.gca().xaxis.set_major_locator(mdates.YearLocator())
fig.set_title("US Economical Data")
axs[0,0].plot(x, gdp.iloc[:,1]);  # Plot some data on the axes.
axs[0,0].set_title("US GDP")
axs[0,1].plot(x2, homeprices.iloc[:,1]);
axs[0,1].set_title("US Home Prices Index")
axs[1,0].plot(x3, sp_500.iloc[:,1]);
axs[1,0].set_title("S&P 500")
axs[1,1].plot(x4, unrate.iloc[:,1]);
axs[1,1].set_title("US Unemployment Rate")
fig.autofmt_xdate()
```
