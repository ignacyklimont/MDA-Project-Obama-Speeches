```python
import fitz
import stanza
import nltk
from textblob import TextBlob

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

%matplotlib inline
```

```python
nlp = stanza.Pipeline(lang='en', processors='tokenize,sentiment',verbose=False)
```

```python
with fitz.open(r"insert file path here") as doc:
    text = ""
    for page in doc:
        text += page.get_text()
```

```python
stop_words = set(stopwords.words('english'))  
df_speech = []
sentences = nltk.sent_tokenize(text)
for sent in sentences:
    word_tokens = word_tokenize(sent)  
    filtered_sentence = [w for w in word_tokens if not w in stop_words]  
    x = ''
    for word in filtered_sentence:
        x +=' '+word
    df_speech.append(x)
```

```python
df = pd.DataFrame(df_speech)
df.columns = ['text']
df['sentiment_stanza'] = ''
df['sentiment_blob'] = ''
```

```python
for idx in df.index:
    doc = nlp(df.loc[idx,'text'])
    for i, sentence in enumerate(doc.sentences):
        df.loc[idx,'sentiment_stanza']=float(sentence.sentiment-1)
```

```python
def blob_sentiment(txt):
    sent = TextBlob(txt).sentiment.polarity
    return sent
```

```python
df['sentiment_blob'] = df['text'].apply(lambda x : blob_sentiment(x))
```
